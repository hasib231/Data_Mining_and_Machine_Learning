```python
grade=["A","B","C","D"]
data = [4,2,1,3]
cols=['r','g','b','m']
expl=[0,0,0.2,0]
textColor={'color':'w'}

plt.pie(data,labels=grade,autopct="%1.1f%%",colors=cols,startangle=90,counterclock=False,explode=expl,shadow=True,textprops=textColor)
plt.show()
```
